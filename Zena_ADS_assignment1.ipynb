# Scatter Plot

```python
def plot_date_vs_temp_with_precip_size(df):
    '''
    This function creates a scatter plot of temperature over a period of 10 years,
    with the marker size representing the precipitation amount and the marker shape
    based on precipitation type.
    
    Parameters:
    - df: DataFrame containing at least 'temp', 'precip', and 'preciptype' columns, 
          and a datetime index representing dates.
          
    Note: Marker size is set to the precipitation amount if greater than 0; otherwise, 
          a minimum size of 10 is used. Marker shape varies by precipitation type.
    '''
    # Define marker and color mapping for each precipitation type
    marker_map = {'rain': 'o', 'rain,snow': 'P', 'None': 's'}
    color_map = {'rain': 'blue', 'rain,snow': 'purple', 'None': 'gray'}
    
    # Loop through each precipitation type and plot with customized marker, color, and size
    for precip_type, marker in marker_map.items():
        subset = df[df['preciptype'] == precip_type]
        
        # Set size for points: a minimum size for 'None' or zero precipitation
        size = subset['precip'].apply(lambda x: x if x > 0 else 10)
        
        plt.scatter(
            subset.index, subset['temp'],  # Using datetime index on x-axis
            s=size,  # Size based on precipitation
            label=precip_type, marker=marker, alpha=0.6,
            color=color_map[precip_type]
        )
    
    # Add axis labels, legend, and title
    plt.xlabel('Date')
    plt.ylabel('Temperature (°C)')
    plt.legend(title='Precipitation Type')
    plt.title('Temperature Over Time with Precipitation as Marker Size')
    plt.xticks(rotation=45)
    plt.tight_layout()  # Adjust layout for readability
    plt.show()
```

```python
plot_date_vs_temp_with_precip_size(df)
```

```python
def plot_temperature_boxplot(df):
    plt.figure(figsize=(10, 6))
    
    # Create a boxplot for the temperature data
    plt.boxplot([df['temp'][df.index.year == year] for year in df.index.year.unique()], 
                labels=df.index.year.unique())
    
    plt.title('Boxplot of Temperature by Year')
    plt.xlabel('Year')
    plt.ylabel('Temperature')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Call the function with your DataFrame
plot_temperature_boxplot(df)
```

# Bar Diagram

```python
def plot_rain_snow_count_by_month_year(df):
    '''
    This function displays a grouped bar chart showing the count of rain and snow occurrences 
    in Pokhara over the last decade, broken down by month and year.
    
    Parameters:
    - df: DataFrame with at least 'preciptype' column indicating 'rain,snow' and a datetime index.
    '''
    # Filter for 'rain,snow' precipitation type
    rain_snow_df = df[df['preciptype'] == 'rain,snow'].copy()
    
    # Extract year and month from the datetime index
    rain_snow_df['year'] = rain_snow_df.index.year
    rain_snow_df['month'] = rain_snow_df.index.month
    
    # Group by year and month to count occurrences, then pivot for easy plotting
    monthly_counts = rain_snow_df.groupby(['year', 'month']).size().unstack(fill_value=0)
    
    # Define bar width and calculate bar positions
    bar_width = 0.15
    years = monthly_counts.index
    num_months = monthly_counts.shape[1]
    
    # Adjust spacing based on number of months to prevent overlap
    bar_positions = np.arange(len(years))
    
    plt.figure(figsize=(14, 7))

    # Plot each month with an offset for grouped bar display
    for i, month in enumerate(monthly_counts.columns):
        month_name = calendar.month_name[month]
        plt.bar(
            bar_positions + i * bar_width,
            monthly_counts[month],
            width=bar_width,
            label=month_name,
            alpha=0.7
        )
    
    # Customize x-axis labels to show each year
    plt.xticks(bar_positions + (num_months * bar_width) / 2, years, rotation=45)
    
    # Add labels, legend, and title
    plt.xlabel('Year')
    plt.ylabel('Count of Rain and Snow Occurrences')
    plt.title('Monthly Counts of Rain and Snow by Year')
    plt.legend(title='Month')
    plt.tight_layout()
    plt.show()
```

```python
# Call the function with your DataFrame
plot_rain_snow_count_by_month_year(df)

```

```python
def plot_confusion_matrix_with_major_temp_threshold(df):
    '''
    This function categorizes temperature into "Lower Temp" and "Higher Temp" based on a threshold,
    and categorizes precipitation into "No Precip", "Lower Precip", and "Higher Precip" categories.
    It then plots a confusion matrix showing the relationship between these categories.
    '''
    # Create new columns with explicit .loc indexing to avoid SettingWithCopyWarning
    df = df.copy()  # Make a copy of the DataFrame to avoid modifying the original
    
    # Categorize based on temperature threshold
    df.loc[:, 'temp_category'] = np.where(df['temp'] < 15, 'Lower Temp', 'Higher Temp')
    
    # Categorize based on precipitation levels
    df.loc[:, 'precip_category'] = np.select(
        [
            df['precip'] == 0,          # No precipitation
            (df['precip'] > 0) & (df['precip'] < 5),  # Lower precipitation,
            (df['precip'] >= 5) & (df['precip'] <= 80), # Medium
            df['precip'] >= 5          # Higher precipitation
        ],
        ['No Precip', 'Lower Precip', 'Medium Precip', 'Higher Precip'],
        default='Unknown'
    )
    
    # Create confusion matrix with cross-tabulation
    confusion_matrix = pd.crosstab(df['temp_category'], df['precip_category'])
    
    # Plot confusion matrix as a heatmap with Seaborn
    plt.figure(figsize=(6, 4))
    sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", cbar_kws={'label': 'Counts'})
    
    # Add labels and title
    plt.xlabel('Precipitation Category')
    plt.ylabel('Temperature Category')
    plt.title('Confusion Matrix for Temperature vs. Precipitation (Based on Major Temp Threshold)')
    plt.tight_layout()
    plt.show()

# Example usage with your DataFrame
plot_confusion_matrix_with_major_temp_threshold(df)
```
